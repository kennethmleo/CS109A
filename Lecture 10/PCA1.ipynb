# PCA Exercise 1

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

## Loading data

```python
X = np.loadtxt('data/PCA1.csv')
```

Standardize your $X$ matrix.

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit(X).transform(X)
```

```python
plt.plot(X[:,0], X[:,1], "+", alpha=0.8)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('X1', fontsize=15)
plt.ylabel('X2', fontsize=15)
plt.grid()
plt.show()
```

## PCA
Transform your $X$ matrix using PCA with 2 components.

```python
### edTest(test_PCA_fit) ###
from sklearn.decomposition import PCA

pca = PCA().fit(X)
X_pca = pca.transform(X)

print("PCA shape:", X_pca.shape)
```

## Visualization

```python
plt.figure(figsize=(6, 5))

plt.plot(X[:,0], X[:,1], "+", alpha=0.8, label='Original')
plt.plot(X_pca[:,0], X_pca[:,1], "+", alpha=0.8, label='Transformed')

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('X1', fontsize=15)
plt.ylabel('X2', fontsize=15)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
