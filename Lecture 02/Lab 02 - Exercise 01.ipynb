## Tracking Covid-19 at U.S. Colleges and Universities

The New York Times is releasing counts of Covid-19 cases reported on college and university campuses in the United States. Since late July, they have been conducting a rolling survey of American colleges and universities — including every four-year public institution and every private college that competes in N.C.A.A. sports — to track the number of coronavirus cases reported among students and employees. The survey now includes more than 1,900 colleges. Starting in 2021 the number of cases in 2021 is also included. 

Colleges and universities that have reported zero cases will be listed with a zero for cases. We have excluded missing values.

### Data

Data can be found in the `colleges.csv` file. 
```
date,state,county,city,ipeds_id,college,cases,cases_2021,notes
2021-02-26,Alabama,Madison,Huntsville,100654,Alabama A&M University,41,,

2021-02-26,Alabama,Jefferson,Birmingham,100663,University of Alabama at Birmingham,2856,570,"Total is known to include one or more cases from a medical school, medical center, teaching hospital, clinical setting or other academic program in health sciences."
```    
The fields have the following definitions:

- **date**: The date of the last update.
- **state**: The state where the college is located.
- **county**: The county where the college is located.
- **city**: The city where the college is located.
- **ipeds_id**: The Integrated Postsecondary Education Data System (IPEDS) ID number for the college.
- **college**: The name of the college or university.
- **cases**: The total number of reported Covid-19 cases among university students and employees in all fields, including those whose roles as doctors, nurses, pharmacists or medical students put them at higher risk of contracting the virus, since the beginning of the pandemic.

#### Plot cases by state/county

````python
cases_by_state = colleges.groupby("state")["cases"].sum().sort_values(ascending=False)
state_with_highest_case = colleges.groupby("state")["cases"].sum().idxmax()
colleges_highest_state_case = colleges[colleges["state"] == state_with_highest_case]
cases_by_county = colleges_highest_state_case.groupby("county")["cases"].sum().sort_values(ascending=False)
````

````python
plt.figure(figsize = (24,6))
plt.bar(cases_by_state.index, cases_by_state.values)
plt.xlabel("state")
plt.ylabel("number of cases")
plt.xticks(rotation=55)
plt.show()

plt.figure(figsize = (24,6))
plt.bar(cases_by_county.index, cases_by_county.values)
plt.xlabel("county in state with highest number of case")
plt.ylabel("number of cases")
plt.xticks(rotation=55)
plt.show()
````

#### Try to visualize COVID cases using geopandas

````python
colleges['location'] = colleges['city'] + ", " + colleges['state']
````

````python
duplicates = colleges['location'][colleges['location'].duplicated()]
print(duplicates)
````

````python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd
from tqdm import tqdm
import time
import os

csv_path = "geocoded_locations.csv"

if os.path.exists(csv_path):
    unique_locations = pd.read_csv(csv_path)
    if 'lat' not in unique_locations.columns:
        unique_locations['lat'] = None
    if 'lon' not in unique_locations.columns:
        unique_locations['lon'] = None
else:
    unique_locations = colleges[['location']].drop_duplicates().copy()
    unique_locations['lat'] = None
    unique_locations['lon'] = None

geolocator = Nominatim(user_agent="college-geocode", timeout=10)
geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=1,
    max_retries=3,
    error_wait_seconds=10.0,
    swallow_exceptions=True
)

for idx in tqdm(unique_locations.index, desc="Geocoding locations"):
    if pd.notnull(unique_locations.at[idx, 'lat']) and pd.notnull(unique_locations.at[idx, 'lon']):
        continue 

    loc_str = unique_locations.at[idx, 'location']
    geo = geocode(loc_str)
    if geo:
        unique_locations.at[idx, 'lat'] = geo.latitude
        unique_locations.at[idx, 'lon'] = geo.longitude

    unique_locations.to_csv(csv_path, index=False)
````

````python
unique_locations.head()
````

````python
colleges_with_coords = colleges.merge(unique_locations, on='location', how='left')
location_cases = colleges_with_coords.groupby(['location', 'lat', 'lon'], as_index=False)['cases'].sum()
location_cases.head()
````

````python
location_cases = location_cases[
    location_cases['lat'].notna() &
    location_cases['lon'].notna() &
    (location_cases['lat'] != 0) &
    (location_cases['lon'] != 0)
]
````

````python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import geodatasets

geometry = [Point(xy) for xy in zip(location_cases['lon'], location_cases['lat'])]
gdf = GeoDataFrame(location_cases, geometry=geometry)   

world = gpd.read_file(geodatasets.data.naturalearth.land['url'])
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=15);
````
